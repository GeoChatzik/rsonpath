```python
import matplotlib.pyplot as plt
import numpy as np
import json
import math

def get_color(n, i):
    x = (i + 1.0) / n
    val = 255.0 - (x * 255.0)
    hex_str = hex(math.floor(val)).split('x')[1]
    return f'#{hex_str}{hex_str}{hex_str}'

def get_colors_for(n):
    return list(map(lambda i: get_color(n, i), range(0, n)))

measurements1 = [
    0.52455,
    0.40287,
    1.16732,
    1.54521
]
measurements2 = [
    0.52502,
    0.40180,
    1.16502,
    1.53705
]

nm1 = np.array(measurements1)
nm2 = np.array(measurements2)
diffs = 1000 * (nm1 - nm2)

names = [
    "recursive, no SIMD",
    "main, no SIMD",
    "recursive",
    "main"
]

fig, ax = plt.subplots(figsize=(7, 5))

colors = get_colors_for(len(diffs))

ax.bar(names, diffs, width=0.8, edgecolor="white", linewidth=0.7, color=colors)

ax.set_yticks(np.arange(-2, 10, 1))

ax.set_ylabel("delta MB/s")

ax.grid(axis = 'y', linestyle = '--')
plt.savefig("plot.svg", format="svg")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import json
import math

def get_color(n, i):
    x = (i + 1.0) / n
    val = 255.0 - (x * 255.0)
    hex_str = hex(math.floor(val)).split('x')[1]
    return f'#{hex_str}{hex_str}{hex_str}'

def get_colors_for(n):
    return list(map(lambda i: get_color(n, i), range(0, n)))

measurements1 = [
    0.51337,
    0.48899,
    0.51726,
    0.50618
]
measurements2 = [
    0.39295,
    0.38165,
    0.40072,
    0.39346
]

measurements3 = [
    1.06060,
    1.07692,
    1.21050,
    1.14493
]

measurements4 = [
    1.46722,
    1.39570,
    1.55828,
    1.49318
]

names = [
    "recursive, no SIMD",
    "main, no SIMD",
    "recursive",
    "main"
]

xs = ["person", "profession", "properties", "combined"]

colors = get_colors_for(4)

fig, ax = plt.subplots(figsize=(10,5))

ax.plot(xs, measurements1, marker='o', color=colors[0], label=names[0])
ax.plot(xs, measurements2, marker='^', color=colors[1], label=names[1])
ax.plot(xs, measurements3, marker='s', color=colors[2], label=names[2])
ax.plot(xs, measurements4, marker='v', color=colors[3], label=names[3])
ax.legend()

ax.set_yticks(np.arange(0.0, 2.0, 0.2))
ax.set_ylabel("GB/s")

ax.grid(axis = 'y', linestyle = '--')
plt.savefig("plot.svg", format="svg")
plt.show()
```
