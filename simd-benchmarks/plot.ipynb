```python
import matplotlib.pyplot as plt
import numpy as np
import json
import math

def get_color(n, i):
    x = (i + 1.0) / n
    val = 255.0 - (x * 255.0)
    hex_str = hex(math.floor(val)).split('x')[1]
    return f'#{hex_str}{hex_str}{hex_str}'

def get_colors_for(measurements):
    return list(map(lambda i: get_color(len(measurements), i), range(0, len(measurements))))

measurements1 = [
    0.43731,
    1.24747
]

measurements2 = [
    0.43460,
    2.15702
]

names = [
    "nosimd",
    "avx2-simd",
]

fig, ax = plt.subplots(ncols=2, sharey=True)

colors1 = get_colors_for(measurements1)
colors2 = get_colors_for(measurements2)

ax[0].bar(names, measurements1, width=0.8, edgecolor="white", linewidth=0.7, color=colors1)
ax[1].bar(names, measurements2, width=0.8, edgecolor="white", linewidth=0.7, color=colors2)

ax[0].set_yticks(np.arange(0.0, 2.5, 0.2))

ax[0].set_title("Compressed")
ax[1].set_title("Prettified")
ax[0].set_xlabel("implementation")
ax[1].set_xlabel("implementation")
ax[0].set_ylabel("GB/s")

ax[0].grid(axis = 'y', linestyle = '--')
ax[1].grid(axis = 'y', linestyle = '--')
plt.savefig("plot.svg", format="svg")
plt.show()
```
