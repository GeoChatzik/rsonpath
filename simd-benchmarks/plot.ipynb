```python
import matplotlib.pyplot as plt
import numpy as np
import json
import math

def get_color(n, i):
    x = (i + 1.0) / n
    val = 255.0 - (x * 255.0)
    hex_str = hex(math.floor(val)).split('x')[1]
    return f'#{hex_str}{hex_str}{hex_str}'

def get_colors_for(n):
    return list(map(lambda i: get_color(n, i), range(0, n)))

measurements1 = [
    1.595,
    1.117,
    1.631,
    1.579,
    1.292,
    1.380,
    1.326,
    0.515,
    0.512
]

measurements2 = [
    7.798,
    6.357,
    5.442,
    3.336,
    1.826,
    1.751,
    1.761,
    1.760,
    1.756
]

measurements3 = [
    10.783,
    9.885,
    8.428,
    5.045,
    2.893,
    2.795,
    1.572,
    1.573,
    1.573
]

names = [
    "nosimd",
    "sse2-simd",
    "avx2-simd",
]

xs = [2, 3, 4, 8, 15, 16, 32, 33, 48]

colors = get_colors_for(3)

fig, ax = plt.subplots(figsize=(10,5))

ax.plot(xs, measurements1, marker='o', color=colors[0], label=names[0])
ax.plot(xs, measurements2, marker='s', color=colors[1], label=names[1])
ax.plot(xs, measurements3, marker='v', color=colors[2], label=names[2])
ax.legend()

ax.set_xticks(np.arange(0.0, 49.0, 4.0))
ax.set_yticks(np.arange(0.0, 13.0, 1.0))
ax.set_ylabel("GB/s")

ax.grid(axis = 'y', linestyle = '--')
plt.savefig("plot.svg", format="svg")
plt.show()
```
